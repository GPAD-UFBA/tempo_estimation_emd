```python
import numpy as np
from upemd import  UPEMD
```

```python
import matplotlib.pyplot as plt 
```

```python
sample_rate = 1000
seconds = 10
num_samples = sample_rate*seconds

time_vect = np.linspace(0, seconds, num_samples)
freq = 5
```

```python
x = np.cos(2 * np.pi * 1 * time_vect)
x += np.sin(2 * np.pi * 2.2e-1 * time_vect)  
```

```python
plt.plot(x)
```

```python
upemd = UPEMD(num_imf=2)
```

```python
imfs = upemd.upemd(x)
```
